## Introduction
The Titanic dataset contains demographic and passenger information for people aboard the Titanic.
This analysis explores patterns related to survival rates and passenger characteristics.

**Dataset Source:** [Titanic Dataset on GitHub](https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv)


## Objectives
- Understand the dataset structure.
- Identify missing values.
- Summarize key statistics.
- Visualize survival rates across different factors.
- Highlight interesting patterns.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv"
df = pd.read_csv(url)
```

```python
# Show first few rows
print("First 5 rows:")
print(df.head())
```

```python
# Show last few rows
print("Last 5 rows:")
print(df.tail())
```

```python
# Dataset info
print("\nDataset Info:")
print(df.info())
```

```python
# Missing values
print("\nMissing values count:")
print(df.isnull().sum())
```

```python
# Basic statistics
print("\nSummary statistics:")
print(df.describe())
```

```python
# Visualization 1 — Survival count
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
# Visualization 2 — Survival by gender
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
# Visualization 3 — Age distribution
plt.figure(figsize=(6,4))
sns.histplot(df['Age'].dropna(), kde=True, bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
# Visualization 4 - Correlation Heat Map
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

## Observations & Insights
- Women had a much higher survival rate than men.
- First-class passengers had better chances of survival compared to third class.
- Younger passengers had slightly higher survival rates.
- Fare positively correlated with survival probability.


## Conclusion
This EDA reveals strong survival patterns based on gender, class, and fare.  
Future steps could include building a predictive model using machine learning.
