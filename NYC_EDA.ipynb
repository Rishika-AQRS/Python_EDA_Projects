### Data cleaning

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

Dropped the redundant index column created during CSV export, as it does not carry analytical value.


```python
df = df[df['trip_distance'] > 0]

```

Removed trips with zero or negative trip distance to ensure meaningful distance and fare analysis.


```python
df = df[df['fare_amount'] > 0]
```

Filtered out trips with zero or negative fare amounts, as these values are invalid for revenue-related insights.


```python
df = df[df['trip_duration'] > 0]
```

Removed records with non-positive trip durations, ensuring all trips represent valid taxi rides.

```python
df['payment_type'] = df['payment_type'].astype('category')
df['store_and_fwd_flag'] = df['store_and_fwd_flag'].astype('category')
```

Converted categorical columns such as payment type and store-and-forward flag to categorical data types for efficient analysis.


```python
df['amount_diff'] = abs(df['total_amount'] - df['calculated_total_amount'])
df = df[df['amount_diff'] < 1]  # tolerance of $1
df.drop(columns=['amount_diff'], inplace=True)
```

Validated total fare consistency by comparing reported and calculated total amounts, removing records with significant discrepancies.


```python
df.shape
```

After cleaning, the dataset retains high-quality trip records suitable for reliable exploratory analysis.

### Anomaly Detection

##### Trip Distance VS Fare (Outlier Focus)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='trip_distance',
    y='fare_amount',
    data=df,
    alpha=0.3, 
    color='brown'
)
plt.title("Trip Distance vs Fare Amount (Outlier Detection)")
plt.xlabel("Trip Distance (miles)")
plt.ylabel("Fare Amount ($)")
plt.show()
```

While fare generally increases with trip distance, several anomalies are observed where short-distance trips have unusually high fares or long-distance trips have comparatively low fares. These may result from tolls, surcharges, route variations, or data inconsistencies.


### Detecting Fare Outliers using IQR

```python
Q1 = df['fare_amount'].quantile(0.25)
Q3 = df['fare_amount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
fare_outliers = df[(df['fare_amount'] < lower_bound) | (df['fare_amount'] > upper_bound)]
fare_outliers.shape
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['fare_amount'])
plt.title("Fare Amount Outliers")
plt.show()
```

The IQR method was used to identify fare anomalies. Trips with fare amounts significantly lower or higher than the normal range were classified as outliers. The box plot highlights extreme fare values that deviate from typical pricing patterns, representing potential anomalies or special-case trips.



### Short Distance but High Fare Trips

```python
high_fare_threshold = df['fare_amount'].quantile(0.95)
short_high_fare = df[
    (df['trip_distance'] < 2) &
    (df['fare_amount'] > high_fare_threshold)
]
short_high_fare.shape
```

Short-distance trips with unusually high fares were identified, suggesting the impact of tolls, minimum fare rules, or surge pricing effects.


### Long Distance but Low Fare Trips

```python
low_fare_threshold = df['fare_amount'].quantile(0.05)
long_low_fare = df[
    (df['trip_distance'] > df['trip_distance'].quantile(0.95)) &
    (df['fare_amount'] < low_fare_threshold)
]
long_low_fare.shape
```

A small number of long-distance trips with relatively low fares were detected, which may indicate discounts, route anomalies, or data recording issues.


### Time-Based Anomalies

```python
off_peak_high_fare = df[
    (df['hour_of_day'].between(1, 4)) &
    (df['fare_amount'] > high_fare_threshold)
]
off_peak_high_fare.shape
```

High-fare trips during off-peak hours were observed, potentially driven by airport rides or night-time surcharges.


### Business Insights and Recommendations

#### Key Business Insights

##### 1. Time Based Demand Patterns
Taxi demand and fare amounts vary significantly across different hours of the day. Early morning hours show higher average fares, likely driven by airport trips and limited taxi availability. Late-night hours tend to have lower average fares, indicating reduced demand.


##### 2. Distance-Free Relationship
Fare amount generally increases with trip distance; however, several anomalies exist where short trips incur high fares and long trips have relatively low fares. These deviations suggest the influence of tolls, surcharges, or dynamic pricing mechanisms.



##### 3. High-Demand and High-Revenue Zones
Analysis based on taxi zone identifiers shows that a limited number of pickup and dropoff zones account for a disproportionately high number of trips and total revenue. These high-activity zones represent key demand and revenue-generating areas, emphasizing their importance for efficient fleet positioning and zone-based pricing strategies.


##### 4. Impact of Anomalous Trips on Revenue
Outlier trips with unusually high or low fares can significantly impact revenue calculations and performance metrics. Identifying and monitoring such anomalies is essential for accurate reporting and fraud detection.


#### Business Recommendations

##### 1. Optimize Taxi Availability by Time
Increase taxi availability during early morning and peak daytime hours to match higher fare potential and demand. Fleet allocation can be dynamically adjusted based on historical time-based demand patterns.


##### 2. Implement Zone-Based Pricing Strategies
Zone-based demand analysis indicates that certain pickup and dropoff location IDs consistently generate higher trip volumes and revenue. These high-activity zones present opportunities for implementing zone-based or dynamic pricing strategies to optimize revenue while maintaining operational efficiency.


##### 3. Monitor Pricing Anomalies
Regular monitoring of anomalous trips can help identify potential billing issues, fraudulent activity, or data recording errors. Automated anomaly detection systems could improve operational transparency.


##### 4. Data-Driven Oprational Planning
Insights from time, distance, and spatial analysis can be leveraged to design more efficient driver shifts, reduce idle time, and improve overall customer satisfaction.
